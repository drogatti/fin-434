```python
import pandas as pd
import numpy as np
import datetime as dt
from datetime import datetime, date, timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import NullFormatter, FixedLocator
```

```python
btc = pd.read_csv('Data/BTC.csv')
btc = btc.rename(columns={'Vol.': 'vol'})
btc.head()
```

```python
def preprocess(x):
    return float(str(x).replace(",", ""))
def preprocessk(x):
    return float(str(x).replace("K", ""))
```

```python
btc.Price = btc.Price.apply(preprocess)
btc.Open = btc.Open.apply(preprocess)
btc.High = btc.Low.apply(preprocess)
btc.Low = btc.Low.apply(preprocess)
btc.vol = (btc.vol.apply(preprocessk))*1000
```

```python
btc.head()
```

```python
#Sort the DF by Date
btc['Date_new'] = pd.to_datetime(btc['Date'])
btc = btc.sort_values(by=['Date_new'], ascending=True)
btc.head()
```

```python
#Create a TS plot of BTC Price over Time
```

```python
plt.plot(btc.Date_new, btc.Low)
```

```python
#Create a much better looking graph
```

```python
fig, btcplot = plt.subplots(1,1, figsize=(10, 7), constrained_layout=True)
plt.plot(btc.Date_new, btc.Low)
#control x-axis
btcplot.xaxis.set_major_locator(mdates.MonthLocator(bymonth=(1, 7)))
btcplot.xaxis.set_minor_locator(mdates.MonthLocator(bymonth=(3, 5, 9, 11)))
#control y-axis
btcplot.yaxis.set_major_locator(FixedLocator(np.arange(0, 75000, 5000)))
btcplot.yaxis.set_minor_locator(FixedLocator(np.arange(0, 75000, 2500)))
#add grid lines
btcplot.grid(True)

#add labels
btcplot.set_ylabel('Price in $')
btcplot.set_xlabel('Date')
btcplot.set_title('BTC Low Price Over Time', loc='center', fontsize='large')
```

### Read in TREX and LUMBER GSVI

```python
trextrend = pd.read_csv('Data/TREXTREND.csv', header=1)
trextrend = trextrend.rename(columns={"NYSE:TREX: (United States)":"trex_term"})

lumbertrend = pd.read_csv('Data/LUMBERTREND.csv', header=1)
lumbertrend = lumbertrend.rename(columns={"lumber: (United States)":"lumber_term"})
```

### Merge TREX and LUMBER GSVI Data

```python
gsvi = trextrend.merge(lumbertrend, on='Week')
gsvi.head()
```

```python
gsvi.dtypes
```

```python
gsvi['Week_new'] = pd.to_datetime(gsvi['Week'])
gsvi.dtypes
```

### Add a new variable that adds 1 day for our merge with stock data

```python
gsvi = gsvi.assign(week_p1=gsvi.Week_new + timedelta(days=1))
gsvi.head()
```

### Plot TREX and LUMBER GSVI on a Time Series plot

```python
fig, gsviplot = plt.subplots(1,1, figsize=(10, 7), constrained_layout=True)
plt.plot(gsvi.Week_new, gsvi.lumber_term, label='Lumber')
plt.plot(gsvi.Week_new, gsvi.trex_term, label='Trex')
#control x-axis
gsviplot.xaxis.set_major_locator(mdates.MonthLocator(bymonth=(1, 7)))
gsviplot.xaxis.set_minor_locator(mdates.MonthLocator(bymonth=(3, 5, 9, 11)))
#control y-axis
gsviplot.yaxis.set_major_locator(FixedLocator(np.arange(0, 100, 10)))
gsviplot.yaxis.set_minor_locator(FixedLocator(np.arange(0, 100, 5)))
#add grid lines
gsviplot.grid(True)

#add labels
gsviplot.set_ylabel('GSVI Level (Out of 100)')
gsviplot.set_xlabel('Week')
gsviplot.set_title('GSVI Terms Over Time', loc='center', fontsize='large')

#add legend
plt.legend(loc='upper left')
```

### Run a correlation

```python
gsvi[['trex_term', 'lumber_term']].corr()
gsvi.corr(numeric_only=True)
```

### Read in Wood Data

```python
wood = pd.read_csv('Data/WOOD.csv')
wood['date_new'] = pd.to_datetime(wood['Date'])
wood.head()
```

### Scale the Vol variable by 1000, i.e. the new variable will be in 000's

```python
wood['Volume'] = wood['Volume']/1000
wood.describe()
```

### Merge WOOD (ETF Data) with GSVI Data

```python
gsviw = gsvi.merge(wood[['date_new', "Volume"]], left_on='week_p1', right_on='date_new')
gsviw.head()
```

### Plot all 3 series on one Time Series Plot

```python
fig, gsviplot = plt.subplots(1,1, figsize=(10, 7), constrained_layout=True)
plt.plot(gsvi.Week_new, gsvi.lumber_term, label='Lumber')
plt.plot(gsvi.Week_new, gsvi.trex_term, label='Trex')
plt.plot(gsviw.Week_new, gsviw.Volume, label='Volume')

#control x-axis
gsviplot.xaxis.set_major_locator(mdates.MonthLocator(bymonth=(1, 7)))
gsviplot.xaxis.set_minor_locator(mdates.MonthLocator(bymonth=(3, 5, 9, 11)))
#control y-axis
gsviplot.yaxis.set_major_locator(FixedLocator(np.arange(0, 100, 10)))
gsviplot.yaxis.set_minor_locator(FixedLocator(np.arange(0, 100, 5)))
#add grid lines
gsviplot.grid(True)

#add labels
gsviplot.set_ylabel('GSVI Level (Out of 100)')
gsviplot.set_xlabel('Week')
gsviplot.set_title('GSVI Terms Over Time', loc='center', fontsize='large')

#add legend
plt.legend(loc='upper left')
```

### Add a secondary y axis to display Volume separately

```python
fig, gsviplot = plt.subplots(1,1, figsize=(10, 7), constrained_layout=True)
gsviplot2 = gsviplot.twinx()
gsviplot.plot(gsvi.Week_new, gsvi.lumber_term, label='Lumber')
gsviplot.plot(gsvi.Week_new, gsvi.trex_term, label='Trex')
gsviplot2.plot(gsviw.Week_new, gsviw.Volume, label='Volume', color='green')

#control x-axis
gsviplot.xaxis.set_major_locator(mdates.MonthLocator(bymonth=(1, 7)))
gsviplot.xaxis.set_minor_locator(mdates.MonthLocator(bymonth=(3, 5, 9, 11)))
#control y-axis
gsviplot.yaxis.set_major_locator(FixedLocator(np.arange(0, 100, 10)))
gsviplot.yaxis.set_minor_locator(FixedLocator(np.arange(0, 100, 5)))
#add grid lines
gsviplot.grid(True)

#add labels
gsviplot.set_ylabel('GSVI Level (Out of 100)')
gsviplot2.set_ylabel('WOOD ETF Vol (000'+"'"+"s)")
gsviplot.set_xlabel('Week')
gsviplot.set_title('GSVI Terms Over Time', loc='center', fontsize='large')

#add legend
gsviplot.legend(loc='upper left')
gsviplot2.legend(loc='upper right')
```
