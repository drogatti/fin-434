### Import PEP and competitor data from Compustat

```python
data = pd.read_csv("gatti_PEP&COMP.csv")
data = data[['fyear', 'tic', 'at', 'ebit', 'revt', 'ni', 'gp', 'teq']]
```

### Create ratios for Profitability analysis

```python
data['npm'] = (data.ni / data.revt) * 100
data['gpm'] = (data.gp / data.revt) * 100
data['bep'] = data.ebit / data['at']
data['avg_at'] = data['at'].mean()
data['roa'] = data.ni / data.avg_at
data['avg_teq'] = data['teq'].mean()
data['roe'] = data.ni / data.avg_teq
```

### Plot each ratio for Pepsi and competitors on **their own** Time Series plot

```python
fig, npmplot = plt.subplots(1,1, figsize=(10, 7), constrained_layout=True)
plt.plot(data[data['tic'] == 'PEP']['fyear'], data[data['tic'] == 'PEP']['npm'], label='Pepsi')
plt.plot(data[data['tic'] == 'KO']['fyear'], data[data['tic'] == 'KO']['npm'], label='Coca-Cola')
plt.plot(data[data['tic'] == 'NSRGY']['fyear'], data[data['tic'] == 'NSRGY']['npm'], label='Nestle')
plt.plot(data[data['tic'] == 'KDP']['fyear'], data[data['tic'] == 'KDP']['npm'], label='Keurig Dr Pepper')
#control x-axis
npmplot.xaxis.set_major_locator(FixedLocator(np.arange(2019, 2024, 1)))
#control y-axis
npmplot.yaxis.set_major_locator(FixedLocator(np.arange(0, 25, 2.5)))
npmplot.yaxis.set_minor_locator(FixedLocator(np.arange(0, 25, 1.25)))
#add grid lines
npmplot.grid(True)

#add labels
npmplot.set_ylabel('Net Profit Margin (%)')
npmplot.set_xlabel('Year')
npmplot.set_title('Net Profit Margin Over Time', loc='center', fontsize='large')

#add legend
plt.legend(loc='upper left')

#save as a jpeg file
plt.savefig("gatti_npm_timeseries.jpg", format='jpg')
```

### Conduct 3-part DuPont analysis for PEP over 5 years and plot on one Time Series plot

```python
data['atr'] = (data.revt / data['at']) * 100
data['flr'] = data['at'] / data.teq
data['roe_dupont'] = data.roe * 100
```

```python
fig, dupontplot = plt.subplots(1,1, figsize=(10, 7), constrained_layout=True)
dupontplot2 = dupontplot.twinx()
dupontplot.plot(data[data['tic'] == 'PEP']['fyear'], data[data['tic'] == 'PEP']['npm'], label='Net Profit Margin')
dupontplot.plot(data[data['tic'] == 'PEP']['fyear'], data[data['tic'] == 'PEP']['atr'], label='Asset Turnover Ratio')
dupontplot2.plot(data[data['tic'] == 'PEP']['fyear'], data[data['tic'] == 'PEP']['flr'], label='Financial Leverage Ratio', color='red')
dupontplot.plot(data[data['tic'] == 'PEP']['fyear'], data[data['tic'] == 'PEP']['roe_dupont'], label='Return on Equity')

#control x-axis
dupontplot.xaxis.set_major_locator(FixedLocator(np.arange(2019, 2024, 1)))
#control y-axis
dupontplot.yaxis.set_major_locator(FixedLocator(np.arange(0, 100, 10)))
dupontplot.yaxis.set_minor_locator(FixedLocator(np.arange(0, 100, 5)))
#add grid lines
dupontplot.grid(True)

#add labels
dupontplot.set_ylabel('Ratio/Margins (%)')
dupontplot2.set_ylabel("Financial Leverage Ratio ($)")
dupontplot.set_xlabel('Year')
dupontplot.set_title('Pepsi DuPont Components Over Time', loc='center', fontsize='large')

#add legend
dupontplot.legend(loc='upper left')
dupontplot2.legend(loc='upper right')

#save as a jpeg file
plt.savefig("gatti_dupont_timeseries.jpg", format='jpg')
```

### Create a histogram of daily stock returns for PEP and KO over last 5 years

```python
#remotely collect Yahoo data
yf.pdr_override()
pep = pdr.get_data_yahoo("PEP", start="2019-05-01", end="2024-05-01")
pep = pep.reset_index()

ko = pdr.get_data_yahoo("KO", start="2019-05-01", end="2024-05-01")
ko = ko.reset_index()
```

```python
pep_ko = pep.merge(ko, on='Date', suffixes=('_pep', '_ko'))
```

```python
#rename columns in df
pep_ko = pep_ko.rename(columns={"Adj Close_pep":"price_pep"})
pep_ko = pep_ko.rename(columns={"Adj Close_ko":"price_ko"})
```

```python
#calculate the percentage change for each series
pep_ko['lag_pep'] = pep_ko['price_pep'].shift(1)
pep_ko['lag_ko'] = pep_ko['price_ko'].shift(1)
```

```python
pep_ko['ret_pep'] = (pep_ko['price_pep'] - pep_ko['lag_pep']) / pep_ko['lag_pep']
pep_ko['ret_ko'] = (pep_ko['price_ko'] - pep_ko['lag_ko']) / pep_ko['lag_ko']
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6), constrained_layout=True)

#plot the daily return for PEP
sns.histplot(pep_ko['ret_pep'], ax=ax1, bins=35, color='blue').set(title="Pepsi Daily Stock Returns Histogram")
ax1.set_xlabel("Daily Stock Return Values (decimals)")
ax1.set_ylabel("Frequency Over Last 5 Years")
#plot the daily return for KO
sns.histplot(pep_ko['ret_ko'], ax=ax2, bins=35, color='red').set(title="Coca-Cola Daily Stock Returns Histogram")
ax2.set_xlabel("Daily Stock Return Values (decimals)")
ax2.set_ylabel("Frequency Over Last 5 Years")
#save as a jpeg file
plt.savefig("gatti_returns_histogram.jpg", format='jpg')
```
