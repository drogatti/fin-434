```python
#Will need these to help with manipulating our data
import numpy as np
import pandas as pd
import random
#we weill use these to get data from yahoo directly
import pandas_datareader as web
from pandas_datareader import data as pdr
import yfinance as yf
#Will use these to plot everything out
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#Use pandas to reformat columns to be displayed with only
#two decimal points... like a currency
pd.options.display.float_format = '{:6.2f}'.format
```

```python
#set initial parameters
InitialPrincipal = 4000
YearsUntilRetirement = 40
YearlyContributionUntilRetirement = 5000
YearsInRetirement = 20
```

### Build Blank Df for simulations

```python
#dataset to store simulation values
data = pd.DataFrame(columns=['Year', 'Principal', 'Contribution', 'Distribution', 'Rate'])
#append to df values for year 1

if (YearsInRetirement > 0) :
    new_row = pd.DataFrame({'Year': 1, 'Principal': InitialPrincipal, 
                            'Contribution': YearlyContributionUntilRetirement, 
                            'Distribution': 0, 'Rate': 0}, index=[0])
    data = pd.concat([data, new_row], ignore_index=True)
if (YearsInRetirement == 0) :
    new_row = pd.DataFrame({'Year': 1, 'Principal': (0.96*InitialPrincipal), 
                            'Contribution': 0, 
                            'Distribution': 0.04*InitialPrincipal, 'Rate': 0}, index=[0])
    data = pd.concat([data, new_row], ignore_index=True)
```

### Collect Yahoo data remotely

```python
yf.pdr_override()
df = pdr.get_data_yahoo('SPY', start="1992-01-01", end="2024-01-31", interval='1mo')
df = df.reset_index()
```

```python
df['date_dt'] = pd.to_datetime(df['Date'])
df['month_of_date'] = df['date_dt'].dt.month
df['Year'] = df['date_dt'].dt.year
df = df[df['month_of_date'] == 1]
#calculate the return (annual)
df['lag_spy'] = df['Adj Close'].shift(1)
df['ret_spy'] = (df['Adj Close'] - df['lag_spy']) / df['lag_spy']
df['Returns'] = df['ret_spy'] * 100
returns = df[1:]
list = returns['Returns'].tolist()
```

### Iterate through years until retirement

```python
for i in range(YearsUntilRetirement - 1) :
    #calculate new year field
    NewYear = i + 2

    #grab a random interest rate from SPY annual rets
    random_return = random.choice(list) / 100

    #Apply the random interest rate to last year's balance and add a contribution
    NewPrincipal = data.loc[i]['Principal'] * (1 + random_return) + YearlyContributionUntilRetirement
    new_row = pd.DataFrame({'Year': NewYear, 'Principal': NewPrincipal, 
                            'Contribution': YearlyContributionUntilRetirement, 
                            'Distribution': 0, 'Rate': random_return}, index=[0])
    data = pd.concat([data, new_row], ignore_index=True)
```

### Iterate through years of retirement

```python
for i in range(YearsInRetirement) :
    #calculate new year field
    NewYear = data.iloc[i + YearsUntilRetirement - 1]['Year'] + 1

    #grab a random interest rate from SPY annual rets
    random_return = random.choice(list) / 100
    #new distribution to the client
    NewDistribution = data.loc[i + YearsUntilRetirement - 1]['Principal'] * (0.04)
    #Apply the random interest rate to last year's balance and add a contribution
    NewPrincipal = .96 * data.loc[i + YearsUntilRetirement - 1]['Principal'] * (1 + random_return)
    new_row = pd.DataFrame({'Year': NewYear, 'Principal': NewPrincipal, 
                            'Contribution': 0, 
                            'Distribution': NewDistribution, 'Rate': random_return}, index=[0])
    data = pd.concat([data, new_row], ignore_index=True)
```

### Build a function for retirement applications

```python
def retire4(InitialPrincipal, YearsUntilRetirement, YearlyContributionUntilRetirement, YearsInRetirement) :
    data = pd.DataFrame(columns=['Year', 'Principal', 'Contribution', 'Distribution', 'Rate'])
#append to df values for year 1

    if (YearsInRetirement > 0) :
        new_row = pd.DataFrame({'Year': 1, 'Principal': InitialPrincipal, 
                            'Contribution': YearlyContributionUntilRetirement, 
                            'Distribution': 0, 'Rate': 0}, index=[0])
        data = pd.concat([data, new_row], ignore_index=True)
    if (YearsInRetirement == 0) :
        new_row = pd.DataFrame({'Year': 1, 'Principal': (0.96*InitialPrincipal), 
                            'Contribution': 0, 
                            'Distribution': 0.04*InitialPrincipal, 'Rate': 0}, index=[0])
        data = pd.concat([data, new_row], ignore_index=True)

    #iterate through years until retire
    for i in range(YearsUntilRetirement - 1) :
    #calculate new year field
        NewYear = i + 2

    #grab a random interest rate from SPY annual rets
        random_return = random.choice(list) / 100

    #Apply the random interest rate to last year's balance and add a contribution
        NewPrincipal = data.loc[i]['Principal'] * (1 + random_return) + YearlyContributionUntilRetirement
        new_row = pd.DataFrame({'Year': NewYear, 'Principal': NewPrincipal, 
                            'Contribution': YearlyContributionUntilRetirement, 
                            'Distribution': 0, 'Rate': random_return}, index=[0])
        data = pd.concat([data, new_row], ignore_index=True)

    #iterate through years in retire
    for i in range(YearsInRetirement) :
        #calculate new year field
        NewYear = data.iloc[i + YearsUntilRetirement - 1]['Year'] + 1

        #grab a random interest rate from SPY annual rets
        random_return = random.choice(list) / 100
        #new distribution to the client
        NewDistribution = data.loc[i + YearsUntilRetirement - 1]['Principal'] * (0.04)
        #Apply the random interest rate to last year's balance and add a contribution
        NewPrincipal = .96 * data.loc[i + YearsUntilRetirement - 1]['Principal'] * (1 + random_return)
        new_row = pd.DataFrame({'Year': NewYear, 'Principal': NewPrincipal, 
                            'Contribution': 0, 
                            'Distribution': NewDistribution, 'Rate': random_return}, index=[0])
        data = pd.concat([data, new_row], ignore_index=True)
    return data
```

```python
data2 = retire4(4000, 40, 5000, 20)
data2.head()
```

```python
sns.set(rc={'figure.figsize': (16, 8)})
sns.set_style('whitegrid')
sns.set_palette('Greens')
plt.ticklabel_format(style='plain', axis='y')

sns.lineplot(x='Year', y="Principal", data=data2, color='red').\
    set_title("4% Rule")

plt.axvline(x=YearsUntilRetirement, color='blue')
```

### Adjust for inflation

```python
simulation100['time'] = simulation100['Year'].astype(float) - 1
simulation100.head()
```

```python
inflation = 0.035
simulation100['inf_adj'] = (1 + inflation)**(simulation100['time'])
simulation100['Principal_adj'] = simulation100['Principal'] / simulation100['inf_adj']
simulation100['Distribution_adj'] = simulation100['Distribution'] / simulation100['inf_adj']
simulation100.head()
```

### Create entire retirement function

```python
def simretire(prin, ytr, cont, yir) :
    simulation100 = pd.DataFrame(columns=['Year', 'Principal', 'Contribution', 'Distribution', 'Rate'])
    for i in range(100) :
        df = retire4(prin, ytr, cont, yir)

        simulation100 = pd.concat([simulation100, df])

    simulation100['time'] = simulation100['Year'].astype(float) - 1
    inflation = 0.035
    simulation100['inf_adj'] = (1 + inflation)**(simulation100['time'])
    simulation100['Principal_adj'] = simulation100['Principal'] / simulation100['inf_adj']
    simulation100['Distribution_adj'] = simulation100['Distribution'] / simulation100['inf_adj']
    sim100 = simulation100
    return sim100
```

```python
def retireplot(prin, work, cont, retire) :
    sim100 = simretire(prin, work, cont, retire)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 8), constrained_layout=True)
    sns.lineplot(x='Year', y="Principal_adj", ax=ax1, data=sim100, color='red', errorbar=('ci', 95)).\
        set_title("4% Rule Simulation Principal")

    ax1.axvline(x=work, color='blue')

    sns.lineplot(x='Year', y="Distribution_adj", ax=ax2, data=sim100, color='green').\
        set_title("4% Rule Simulation Distribution")

    ax2.axvline(x=work, color='red')
```
