### Allow the User to input Tickers

```python
tic1 = str(input('Enter Ticker in CAPS'))
tic2 = str(input('Enter Ticker in CAPS'))
```

### Add a label to be used for graphs later

```python
label = tic1+' and '+ tic2 + ' over time'
label2 = tic1+' moving averages'
```

### Remotely collect Yahoo data

```python
yf.pdr_override()
df = pdr.get_data_yahoo(tic1, start="2015-01-01", end="2024-01-31")
df = df.reset_index()

df1 = pdr.get_data_yahoo(tic2, start="2015-01-01", end="2024-01-31")
df1 = df1.reset_index()
```

```python
df2 = df.merge(df1, on='Date', suffixes=('_tic1', '_tic2'))
```

### rename columns in df

```python
stocks = df2.rename(columns={"Adj Close_tic1":"price_tic1"})
stocks = stocks.rename(columns={"Adj Close_tic2":"price_tic2"})
stocks.columns
```

### Calculate the percentage change for each series

```python
stocks['lag_tic1'] = stocks['price_tic1'].shift(1)
stocks['lag_tic2'] = stocks['price_tic2'].shift(1)
```

```python
stocks['ret_tic1'] = (stocks['price_tic1'] - stocks['lag_tic1']) / stocks['lag_tic1']
stocks['ret_tic2'] = (stocks['price_tic2'] - stocks['lag_tic2']) / stocks['lag_tic2']
```

### Calculate the value of $1 invested overtime

```python
stocks['tic1_one'] = stocks['ret_tic1'] + 1
stocks['dol_val_tic1'] = np.cumprod(stocks['tic1_one'])

stocks['tic2_one'] = stocks['ret_tic2'] + 1
stocks['dol_val_tic2'] = np.cumprod(stocks['tic2_one'])
```

### Plot the value of $1 invested

```python
fig, ticplot = plt.subplots(1, 1, figsize=(10, 7), constrained_layout=True)
#add a line to create a second axis
ticplot2 = ticplot.twinx()

#plot
ticplot.plot(stocks.Date, stocks.dol_val_tic1, label="%s"%(str(tic1)), color='green')
ticplot2.plot(stocks.Date, stocks.dol_val_tic2, label="%s"%(str(tic2)), color='orange')

#add grid lines
ticplot.grid(True)

#add axis labels
ticplot.set_ylabel(r"%s in $"%(str(tic1)))
ticplot2.set_ylabel(r"%s in $"%(str(tic2)))
ticplot.set_xlabel("Date")

#set a title
ticplot.set_title('%s'%(str(label)), loc='center', fontsize='large')

#add a legend
ticplot.legend(loc='upper left')
ticplot2.legend(loc='upper right')
```

### Calculate moving averages

```python
df_df = stocks.price_tic1
stocks['rolling_mean_30'] = df_df.rolling(window=30).mean()
stocks['rolling_mean_100'] = df_df.rolling(window=100).mean()
```

### Generate random samples

```python
trials = 1000
samplesize = 100
```

```python
g = np.zeros(trials)
sk = np.zeros(trials)
tm = np.zeros(trials)
me = np.zeros(trials)
```

```python
m = snip['ret_tic1'].mean()
s = snip['ret_tic1'].std()
```

```python
i = 0
while i < trials :
    sample = np.random.normal(m,s,samplesize)
    geomean = st.geometric_mean(sample + 1)
    tm[i] = mstats.tmean(sample, (-0.01, 0.01))
    me[i] = sample.mean()
    sk[i] = stats.skew(sample)
    g[i] = (geomean - 1)
    
    i += 1
```

### Plot the Summary stats from the random loop with seperate graphs

```python
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 6), constrained_layout=True)

#plot the geomean
sns.histplot(g, ax=ax1, bins=25, color='blue').set(title="Geo Mean Histogram")
ax1.set_xlabel("Geo Mean Values")
ax1.set_ylabel("Freq out of "+str(trials)+" Trials")
#plot the skewness
sns.histplot(sk, ax=ax2, bins=25, color='green').set(title="Skewness Histogram")
ax2.set_xlabel("Skewness Values")
ax2.set_ylabel("Freq out of "+str(trials)+" Trials")
#plot the trimmed mean
sns.histplot(tm, ax=ax3, bins=25, color='orange').set(title="Trimmed Mean Histogram")
ax3.set_xlabel("Trimmed Mean Values")
ax3.set_ylabel("Freq out of "+str(trials)+" Trials")
#plot the mean
sns.histplot(me, ax=ax4, bins=25, color='purple').set(title="Mean Histogram")
ax4.set_xlabel("Mean Values")
ax4.set_ylabel("Freq out of "+str(trials)+" Trials")
```

### Create descriptive stats

```python
snip_stats = pd.DataFrame({'Geo Mean Stats':g, 'Skewness Stats':sk})
dfstat = snip_stats.describe()
```
