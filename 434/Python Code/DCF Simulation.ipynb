```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

### Import TSLA data from Compustat

```python
tsla = pd.read_csv('Data/tsla_dcf.csv')
tsla = tsla[['fyear', 'act', 'ebit', 'ebitda', 'lct', 'revt', 'ch', 'dt']]
```

### Create ratios for DCF model

```python
tsla['ebit_sales'] = tsla.ebit / tsla.revt
tsla['ebitda_sales'] = tsla.ebitda / tsla.revt
tsla['nwc'] = tsla.act - tsla.lct
tsla['nwc_sales'] = tsla.nwc / tsla.revt
tsla['dep'] = tsla.ebitda - tsla.ebit
tsla['dep_sales'] = tsla.dep / tsla.revt
```

#### Calculate revenue growth from one period to next

```python
tsla['lag_rev'] = tsla['revt'].shift(1)
tsla['rev_growth'] = (tsla['revt'] - tsla['lag_rev']) / tsla['lag_rev']
tsla.head()
```

### Forecast sales over the next 5 years

```python
years = ['2023A', '2024P', '2025P', '2026P', '2027P', '2028P']
sales = pd.Series(index=years)
sales['2023A'] = tsla.loc[4, 'revt']
```

### Generate a list of sales growth rates

```python
rev_growth_avg = (tsla.loc[4, 'rev_growth']+tsla.loc[3, 'rev_growth']+tsla.loc[2, 'rev_growth'])/3
rev_growth_avg
```

```python
g = np.zeros(6)
i = 0
while i < 6 :
    if i < 4 :
        g[i] = rev_growth_avg
    else :
        g[i] = tsla.loc[4, 'rev_growth'] / (i)

    i += 1
g.tolist()
```

```python
sales[1] = sales[0] * (1 + g[0])
sales[2] = sales[1] * (1 + g[1])
sales[3] = sales[2] * (1 + g[2])
sales[4] = sales[3] * (1 + g[3])
sales[5] = sales[4] * (1 + g[4])
print(sales)
```

### Calculate FCFs w/o Leverage

```python
ebitda_margin = tsla.loc[4, 'ebitda_sales']
depr_percent = tsla.loc[4, 'dep_sales']
ebitda = sales * ebitda_margin
depreciation = sales * depr_percent
nwc_percent = tsla.loc[4, 'nwc_sales']
nwc = sales * nwc_percent
change_in_nwc = nwc.shift(1) - nwc
capex_percent = depr_percent
capex = -(sales * capex_percent)
#we are hard coding the tax rate
tax_rate = .21
ebit = ebitda - depreciation
tax_payment = -ebit * tax_rate
tax_payment = tax_payment.apply(lambda x: min(x, 0))
free_cash_flow = ebit + depreciation + tax_payment + capex + change_in_nwc
```

```python
print(free_cash_flow)
```

### Write data to DF

```python
output = pd.DataFrame([sales, ebit, free_cash_flow], index = ['Sales', 'EBIT', 'FCF']).round(2)
print(output)
```

### Compute the DCF value

```python
#Assumptions Start
cost_of_capital = 0.085
terminal_growth = 0.045
#Assumptions End

terminal_value = ((free_cash_flow[-1] * (1 + terminal_growth)) / (cost_of_capital - terminal_growth))
discount_factor = [(1 / (1 + cost_of_capital))**i for i in range(1, 6)]
dcf_value = (sum(free_cash_flow[1:] * discount_factor) + terminal_value * discount_factor[-1])
```

### Build Monte Carlo Simulation for the DCF

```python
trials = eval(input("How Many Trials? (in int form)"))
```

```python
ebitda_margin_avg = (tsla.loc[4, 'ebitda_sales']+tsla.loc[3, 'ebitda_sales']+tsla.loc[2, 'ebitda_sales'])/3
ebitda_margin_avg
```

```python
nwc_percent_avg = (tsla.loc[4, 'nwc_sales']+tsla.loc[3, 'nwc_sales']+tsla.loc[2, 'nwc_sales'])/3
nwc_percent_avg
```

### Create the Simulation Function

```python
def run_mcs() :
    #create the prob dists
    sales_growth_dist = np.random.normal(rev_growth_avg, scale=0.10, size=trials)
    ebitda_margin_dist = np.random.normal(ebitda_margin_avg, scale=0.05, size=trials)
    nwc_percent_dist = np.random.normal(nwc_percent_avg, scale=0.10, size=trials)
    t_growth_dist = np.random.normal(0.045, scale=0.005, size=trials)

    #calculate the DCF value for each set of random inputs
    output_distribution = []
    for i in range(trials) :
        for year in range(1, 6) :
            sales[year] = sales[year-1]*(1+sales_growth_dist[i])
        ebitda = sales * ebitda_margin_dist[i]
        depreciation = (sales * depr_percent)
        ebit = ebitda - depreciation
        nwc = sales * nwc_percent_dist[i]
        change_in_nwc = nwc.shift(1) - nwc
        capex = -(sales * capex_percent)
        tax_payment = -ebit * tax_rate
        tax_payment = tax_payment.apply(lambda x: min(x, 0))
        
        free_cash_flow = ebit + depreciation + tax_payment + capex + change_in_nwc
        #DCF
        terminal_value = ((free_cash_flow[-1] * (1 + t_growth_dist[i])) / (cost_of_capital - t_growth_dist[i]))
        free_cash_flow[-1] += terminal_value
        discount_factors = [(1 / (1 + cost_of_capital))**i for i in range(1, 6)]
        dcf_value = (sum(free_cash_flow[1:] * discount_factors))
        equity_val = dcf_value - debt + cash
        stock_price = (equity_val * 1000) / shrout
        output_distribution.append(stock_price)
    return output_distribution
```

```python
dcf_value = run_mcs()
```

### Create a user input for stock price

```python
current_pc = eval(input('What is the Current price?'))
```

### Visualize the MCSIM

```python
#calculate the sell % for the graph
sell_pct = (len([x for x in dcf_value if x < (current_pc * 0.8)]) / len(dcf_value))
s = str(round((sell_pct * 100), 2))
#calculate the sell % for the graph
buy_pct = (len([x for x in dcf_value if x > (current_pc * 1.2)]) / len(dcf_value))
b = str(round((buy_pct * 100), 2))
#calculate the hold % for the graph
h = str(round(((1 - sell_pct - buy_pct) * 100), 2))
```

### Plot the histogram

```python
cmap = plt.get_cmap('nipy_spectral')
sell = cmap(0.9)
hold = cmap(0.7)
buy = cmap(0.5)

_,_, bars = plt.hist(dcf_value, bins=50)
for bar in bars :
    if bar.get_x() > current_pc * 1.2 :
        bar.set_facecolor(buy)
    elif bar.get_x() < current_pc * 0.8 :
        bar.set_facecolor(sell)
    else :
        bar.set_facecolor(hold)

#add labels
plt.xlabel("Stock Price [$]")
plt.ylabel("Freq")
plt.title("DCF of TSLA")
#add a line for current price
plt.axvline(x=current_pc, linestyle = '--', linewidth = 2, color='black', label='Current Price')
#create a legend
handles = [Rectangle((0, 0), 0, 0, color=c, ec='k') for c in [sell, hold, buy]]
labels = ["Sell", "Hold", "Buy"]
plt.legend(handles, labels)
#add the text for current price
plt.text(current_pc, 800, ' <------- Current Price', rotation=0)

#add text for % buy sell hold
plt.text(1000, 600, s+'% Sells')
plt.text(1000, 550, b+'% Buys')
plt.text(1000, 500, h+'% Holds')
```
