# Read in Google GSVI data CSV

```python
gsvi = pd.read_csv('Data/tslatrends.csv', header=1)
gsvi.head()
```

### Change variable name

```python
gsvi = gsvi.rename(columns={"NASDAQ:TSLA: (United States)": "GSVI_TSLA"})
gsvi.head()
```

```python
gsvi.dtypes
```

```python
ret['date'] = pd.to_datetime(ret['date'])
gsvi['Week'] = pd.to_datetime(gsvi['Week'])
gsvi['Week'] = gsvi['Week'] + pd.Timedelta('1 day')
```

```python
gsvi.dtypes
```

```python
merge = pd.merge(gsvi, ret, how='left', left_on='Week', right_on='date')
```

```python
merge.head(-1)
```

```python
# Drop observations with missing OPENPRC
merge=merge[merge['OPENPRC'].notna()]
merge.head()
```

### Add a split adjustment to stock data

```python
merge['split_adj']=1
merge['split_adj']=np.where(merge["date"] <= '2022-08-25', 3, merge['split_adj'])
merge['split_adj']=np.where(merge["date"] <= '2020-08-31', 15, merge['split_adj'])
merge['adj_price']=merge['OPENPRC']/merge['split_adj']
```

### Calculate weekly stock returns

```python
merge['lead_prc'] = merge['adj_price'].shift(1)
merge['adj_ret'] = (merge['adj_price'] - merge['lead_prc']) / merge['lead_prc']
merge.head()
```

```python
print(len(merge))
merge=merge[merge['adj_ret'].notna()]
print(len(merge))
merge=merge[merge['GSVI_TSLA'].notna()]
print(len(merge))
```

### Run a regression

```python
#demo how to snip a df based on a single variable condition
merge1 = merge[merge['date'] <= '2022-01-01']
```

```python
formula = 'adj_ret ~ GSVI_TSLA'
results = ols(formula, merge1).fit()
```

```python
print(results.summary())
```

```python
trex = pd.read_csv('Data/TREX.csv')
trex['sun_date'] = pd.to_datetime(trex['Date'])+timedelta(days=-1)
#trex.dtypes
```

### create our own return variables

```python
trex = trex.assign(ret=(trex.Close - trex.Open) / trex.Open)
#trex.columns
```

```python
trex.describe()
```

### Import GSVI for "TREX" and "LUMBER"

```python
trextrend = pd.read_csv('Data/TREXTREND.csv', header = 1)
trextrend = trextrend.rename(columns={"NYSE:TREX: (United States)":"trex_term"})
#trextrend.head()
```

```python
lumbertrend = pd.read_csv('Data/LUMBERTREND.csv', header = 1)
lumbertrend = lumbertrend.rename(columns={"lumber: (United States)":"lumber_term"})
#lumbertrend.head()
```

### Merge the two GSVI DFs

```python
gsvi=trextrend.merge(lumbertrend, left_on='Week', right_on='Week')
gsvi['Week'] = pd.to_datetime(gsvi['Week'])
gsvi.head()
```

```python
gsvi.dtypes
```

### Merge with yahoo data

```python
trex_master = pd.merge(gsvi, trex, how='left', left_on='Week', right_on='sun_date')
trex_master.head()
```

### Run another regression

```python
formula = 'Volume ~ trex_term'
results = ols(formula, trex_master).fit()
```
