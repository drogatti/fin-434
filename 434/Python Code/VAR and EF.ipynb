```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from pandas_datareader import data as pdr
import yfinance as yf
```

### Create a portfolio of assets

```python
assets = ['NVDA', 'TREX', 'IBM', 'HON']
pf_data = pd.DataFrame()
yf.pdr_override()
```

```python
type(assets)
```

```python
for x in assets :
    pf_data[x] = pdr.get_data_yahoo(x, start='2012-8-30', end='2024-1-31')['Adj Close']
```

### Calculate returns and annualize (simple)

```python
returns = pf_data.pct_change()
returns.head(-1)
```

```python
returns.mean() * 252
```

### Calculate Covariance and Correlation

```python
returns.cov() * 252
```

```python
returns.corr()
```

### Construct the efficient frontier

```python
#create random weight for portfolios
portfolio_returns = []
portfolio_volatilities = []
portfolio_weights = []

for x in range(10000) :
    weights = np.random.random(len(assets))
    weights /= np.sum(weights)
    portfolio_weights.append(weights)
    portfolio_returns.append(np.sum(weights*returns.mean()*252))
    portfolio_volatilities.append(np.sqrt(np.dot(weights.T, np.dot(returns.cov()*252, weights))))

portfolio_returns = np.array(portfolio_returns)
portfolio_volatilities = np.array(portfolio_volatilities)
```

```python
len(portfolio_volatilities)
```

```python
data = {'Return': portfolio_returns, 'Volatility': portfolio_volatilities}
for counter, symbol in enumerate(returns.columns.tolist()) :
    data[symbol + '_weight'] = [w[counter] for w in portfolio_weights]
portfolios = pd.DataFrame(data)
portfolios.head()
```

### Plot the EF

```python
portfolios.plot(x='Volatility', y='Return', kind='scatter', figsize=(15, 10));
plt.xlabel('Exp Volatility')
plt.ylabel('Exp Return')
```

### Find the min variance port

```python
min_vol = portfolios.iloc[portfolios['Volatility'].idxmin()]
min_vol
```

### Find highest sharpe ratio

```python
rf = 0.01
portfolios['Sharpe'] = (portfolios['Return'] - rf) / portfolios['Volatility']
portfolios.sort_values(by=['Sharpe'])
```

### Extract optimal risky portfolio

```python
optimal_risky = portfolios.iloc[((portfolios['Return']-rf)/portfolios['Volatility']).idxmax()]
optimal_risky
```

### Plot the Ef w Min Var and Optimal Risky indicated

```python
plt.subplots(figsize=(8, 8))
plt.scatter(portfolios['Volatility'], portfolios['Return'], marker='o', color='green');
plt.scatter(min_vol[1], min_vol[0], color='blue', marker='*', s=200)
plt.scatter(optimal_risky[1], optimal_risky[0], color='yellow', marker='*', s=200)
plt.xlabel('Exp Volatility')
plt.ylabel('Exp Return')
```

### Extract the weights from the optimal risky portfolio

```python
optimal_risky.iloc[2]+optimal_risky.iloc[3]+optimal_risky.iloc[4]+optimal_risky.iloc[5]
```

```python
weights = np.array([optimal_risky.iloc[2],optimal_risky.iloc[3],optimal_risky.iloc[4],optimal_risky.iloc[5]])
weights
```

### Calculate the VAR

```python
#set initial investment
initial_investment = 50000
#calculate returns for each stock
avg_ret = returns.mean()
#calculate portfolio avg returns
port_mean = avg_ret.dot(weights)
#calculate the std dev of portfolio
port_stdev = np.sqrt(weights.T.dot(returns.cov()).dot(weights))
#mean investment
mean_investment = (1 + port_mean) * initial_investment
#calculate the std dev of investment
stdev_investment = initial_investment * port_stdev
```

### Now we calculate VAR

```python
conf_level = 0.01
```

```python
cutoff = norm.ppf(conf_level, mean_investment, stdev_investment)
var_1d = initial_investment - cutoff
var_1d
```

### Calculate n-day VAR

```python
var_array = []
num_days = 77
for x in range(1, num_days+1) :
    var_array.append(np.round(var_1d * np.sqrt(x), 2))
```

### Plot the N-Day VAR

```python
plt.plot(var_array, "r")
plt.xlabel("Day #")
plt.ylabel("Max Portfolio Loss (USD)")
plt.title("Max Portfolio Loss (VAR) over 77-day Window")
```
